#### Handling missing data
 Replacing the blank value in Revenue by meann value

```python
movieData['Revenue (Millions)']=movieData['Revenue (Millions)'].fillna(movieData['Revenue (Millions)'].mean())
```

```python
movieData['Metascore']=movieData['Metascore'].fillna(movieData['Metascore'].mode()[0])
```

Question-1 Which genre movie has collected highest revenue?

```python
movieData['Revenue (Millions)'].idxmax()
```

```python
movieData.iloc[50,2]
```

###Observation

Action, Adventue and Fantasy is the genre that attracts a lot of audience

Question 2 - Did movie of the director affected the votes?

```python
figure = plt.figure(figsize=[20, 7])
movieData.groupby(by='Director').sum()['Votes'].plot.bar(color='green')
plt.show()
```

Observation: 
The votes did varied with the director
Few directors got very high votes and few got very less comparitively.

Question 3-What is the relation between Rating, revenue and votes

```python
# Extracting a subset of features from the dataset
subset = movieData[['Rank', 'Revenue (Millions)', 'Votes']]

# Using magic of pandas area() function for plotting stack area plot
sns.pairplot(data=subset, height=2.5, diag_kind='hist')

plt.title('My Pair Plot')
# Dispaly the output by rendering visual on the screen
plt.show()
```

Observation:
Movies with very high Rank did not collect much revenue except for few outliners
Movies with high rating were also voted less
Movies with lless revenue got less votes


Question 4 -Do movies having high imdb rating earn more revenue?

```python
fig = plt.figure(figsize=(15, 7))
sns.regplot(x='Revenue (Millions)', y='Rating', data=movieData,color='brown')
plt.xticks(ticks=np.arange(0, 2, .10), size=12)
plt.yticks(ticks=np.arange(0, 10, 1), size=12)
plt.xlabel(xlabel='Revenue',size=14)
plt.ylabel(ylabel='Ratings',size=14)
plt.title(label='Revenue Vs Ratings', size=16)
plt.show()
```

Question 5- How many movies were released each year

```python
figure = plt.figure(figsize=[15, 7])
movieData.groupby(by='Year')['Year'].count().plot.bar(color='grey')
plt.show()
```

Question 6:  How many people voted each year

```python


grp_data = movieData.groupby('Year').sum()

figure = plt.figure(figsize=[15, 7])
grp_data['Votes'].plot.bar()

plt.show()   
```

Observation:
Number of votes increased from 2006-2014 , but afterwards it dropped

Question 7: By how many votes did first movie ranked got

```python
movieData[movieData['Rank'] == 1]['Votes'].values
movieData[movieData['Rank'] == 1]['Votes'].values[0] - movieData[movieData['Rank'] == 2]['Votes'].values[0]
```

Observation:
The first ranked movie was far ahead in terms of revenue than Second ranked movie. There was no close competition between both movies.


###Observation:
Generally movies with high IMDB rating have failed to collect high revenue

Question 8-What is the distribution of Metascore?

```python
fig = plt.figure(figsize=(15, 7))
sns.distplot(a=movieData['Metascore'], kde=True)
plt.show()
```

```python
Observation:
Majority of the movies scored metascore of 50-60
The metascore is nomally distributed
```

Question 9: What is the density of Runtime

```python
fig = plt.figure(figsize=(15, 7))
sns.distplot(a=movieData['Runtime (Minutes)'], kde=True)
plt.title(label='Density Distribution of Runtime', size=16)
plt.xlabel(xlabel='Runtime', size=14)
plt.ylabel(ylabel='Density', size=14)
plt.xticks(size=12)
plt.yticks(size=12)
plt.grid(b=True)

plt.show()
```

Observation:
Majority of movies have run time between 100-120 minutes
